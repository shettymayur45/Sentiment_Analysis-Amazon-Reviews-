# Working on Amazon reveiw.

```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/drive/MyDrive/Data Science/Dataset/NLP/amazonreviews.tsv',sep='\t')
```

```python
df
```

```python
df['label'].value_counts()
```

```python
df['review'][0]
```

```python
review1=df.loc[0,'review']
```

```python
sia.polarity_scores(review1)
```

```python
df['scores'] = df['review'].apply(lambda r :sia.polarity_scores(r))
```

```python
df
```

```python
df['compound']=df['scores'].apply(lambda c:c['compound'])
```

```python
df
```

# Creating pred label

```python
df['pred_label']=df['compound'].apply(lambda m:'pos' if m>0 else 'neg')
```

```python
df
```

```python
from sklearn.metrics import classification_report,accuracy_score
```

```python
print(accuracy_score(df['label'],df['pred_label']))
```

```python
print(classification_report(df['label'],df['pred_label']))
```
